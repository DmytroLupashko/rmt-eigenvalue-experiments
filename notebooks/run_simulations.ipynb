```python
# Imports
from pathlib import Path
from typing import Callable

import numpy as np

from src import SimulationConfig, EigenvalueSimulator, BernoulliEnsemble, NormalizedBernoulliEnsemble, GaussianEnsemble, \
    UniformEnsemble
```

```python
# Working functions
def linear_scaling(a: float, b: float) -> Callable[[int], float]:
    def p_func(n: int) -> float:
            return a * n - b

    p_func.__name__ = f"linear_{a}n_-_{b}"
    return p_func

def id(n: int) -> float:
    return n

def power_scaling(beta: float) -> Callable[[int], float]:
    def p_func(n: int) -> float:
        return n ** beta

    p_func.__name__ = f"power_{beta}"
    return p_func


def log(n: int) -> float:
    return np.log(n)


def loglog(n: int) -> float:
    return np.log1p(n)

def constant(p: float) -> Callable[[int], float]:
    def p_func(n: int) -> float:
        return p

    p_func.__name__ = f"constant_{p}"
    return p_func
```

```python
CSV_PATH  = "../data/simulation_results.csv"

# Help function for visualization
def simulate_bernoulli(p_func: Callable[[int], float], start: int, end: int, step: int) -> None:
    for n in range(start, end, step):
        config = SimulationConfig(
            n=n,
            ensemble=BernoulliEnsemble(p_func_name=p_func.__name__, p=p_func(n) / n),
            epsilon=0.05,
            min_samples=0,
            batch_size=5000,
            sweep_tag=f"bernoulli_{p_func.__name__} for [{start}, {end}]"
        )

        print(f"Launching simulation  sweep={config.sweep_tag}  n={config.n}")

        EigenvalueSimulator(config).run(
            log_to=Path("../data/simulation_results.csv"),
        )

def simulate_normalized_bernoulli(p_func: Callable[[int], float], start: int, end: int, step: int) -> None:
    for n in range(start, end, step):
        config = SimulationConfig(
            n=n,
            ensemble=NormalizedBernoulliEnsemble(p_func_name=p_func.__name__, p=p_func(n) / n),
            epsilon=0.05,
            min_samples=0,
            batch_size=5000,
            sweep_tag=f"normalized_bernoulli_{p_func.__name__} for [{start}, {end}]"
        )

        print(f"Launching simulation  sweep={config.sweep_tag}  n={config.n}")

        EigenvalueSimulator(config).run(
            log_to=Path("../data/simulation_results.csv"),
        )

def simulate_normal(start: int, end: int, step: int) -> None:
    for n in range(start, end, step):
        config = SimulationConfig(
            n=n,
            ensemble=GaussianEnsemble(),
            epsilon=0.05,
            min_samples=0,
            batch_size=5000,
             sweep_tag=f"normal for [{start}, {end}]"
        )

        print(f"Launching simulation  sweep={config.sweep_tag}  n={config.n}")
        EigenvalueSimulator(config).run(
            log_to=Path("../data/simulation_results.csv"),
        )

def simulate_uniform(start: int, end: int, step: int) -> None:
    for n in range(start, end, step):
        config = SimulationConfig(
            n=n,
            ensemble=UniformEnsemble(),
            epsilon=0.05,
            min_samples=0,
            batch_size=5000,
            sweep_tag=f"uniform for [{start}, {end}]"
        )

        print(f"Launching simulation  sweep={config.sweep_tag}  n={config.n}")
        EigenvalueSimulator(config).run(
            log_to=Path("../data/simulation_results.csv"),
        )
```

```python
# Run simulations
START = 200
END = 500
STEP = 25
```

```python
simulate_bernoulli(log, start=START, end=END, step=STEP)
simulate_bernoulli(loglog, start=START, end=END, step=STEP)
for b in range(1, 10):
    simulate_bernoulli(power_scaling(b / 10), start=START, end=END, step=STEP)
# Assuming that n are big, b does not make impact
for a in range(1, 10):
    simulate_bernoulli(linear_scaling(a / 10, 0), start=START, end=END, step=STEP)
for p in range(1, 10):
    simulate_bernoulli(constant(p / 10), start=START, end=END, step=STEP)
```

```python
simulate_normalized_bernoulli(log, start=START, end=END, step=STEP)
simulate_normalized_bernoulli(loglog, start=START, end=END, step=STEP)
for b in range(1, 10):
    simulate_normalized_bernoulli(power_scaling(b / 10), start=START, end=END, step=STEP)
# Assuming that n are big, b does not make impact
for a in range(1, 10):
    simulate_normalized_bernoulli(linear_scaling(a / 10, 0), start=START, end=END, step=STEP)
for p in range(1, 10):
    simulate_normalized_bernoulli(constant(p / 10), start=START, end=END, step=STEP)
```

```python
simulate_normal(start=START, end=END, step=STEP)
simulate_uniform(start=START, end=END, step=STEP)
```

```python
simulate_bernoulli(id, start=START, end=END, step=STEP)
simulate_normalized_bernoulli(id, start=START, end=END, step=STEP)
```
